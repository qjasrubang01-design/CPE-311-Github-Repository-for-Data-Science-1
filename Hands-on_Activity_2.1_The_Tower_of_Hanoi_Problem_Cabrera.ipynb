<h3><b>Discussion</b></h3>

The Tower of Hanoi is a mathematical puzzle invented by the French mathematician Edouard LucasLinks to an external site. in 1883.

There are three pegs, source(A), Auxiliary (B) and Destination(C). Peg A contains a set of disks stacked to resemble a tower, with the largest disk at the bottom and the smallest disk at the top. figure 1 Illustrate the initial configuration of the pegs for 3 disks. The objective is to transfer the entire tower of disks in peg A to peg C maintaining the same order of the disks.

Obeying the following rules:

1. Only one disk can be transfer at a time.
2. Each move consists of taking the upper disk from one of the peg and placing it on the top of another peg i.e. a disk can only be moved if it is the uppermost disk of the peg.
3. Never a larger disk is placed on a smaller disk during the transfer.
4. The solution to the puzzle calls for an application of recursive functions and recurrence relations.





Figure 1.1 Initial State of the Tower of Hanoi Problem

A skeletal recursive procedure (Outline) for the solution of the problem for N number of disks is as follows:

1. Move the top N-1 disks from peg A to peg B (using C as an auxiliarypeg)
2. Move the bottom disk from peg A to peg C
3. Move N-1 disks from Peg B to Peg C (using Peg A as an auxiliary peg)

<h3><b>Procedure</b></h3>

1. Implement the algorithm for the Tower of Hanoi Problem as shown above.
2. Write your observations for your written algorithm/solution.

```python
disks=int(input("Please enter number of disks:"))
A = list(range(disks, 0, -1))  
B = [] 
C = []

def hanoi(n, source, helper, target):#Puzzle State
    if n == 1:
        target.append(source.pop())
        print("A:", A, "B:", B, "C:", C)
        return

    hanoi(n-1, source, target, helper)
    target.append(source.pop())
    print("A:", A, "B:", B, "C:", C)
    hanoi(n-1, helper, source, target)

def tower (n, source, des, aux):#Moves
    if n==0:
        return 0
    moves = 0
    moves+=tower(n - 1, source, aux, des)
    print("Move disk", n , "from tower", source, "to tower", des)
    moves+=1
    moves+=tower(n - 1, aux, des, source)
    return moves

print("Disks entered:", disks)
if disks<=0:
    print("Please put a number greater than 0")
else:
    print("\nPuzzle Moves\n")
    x=tower(disks, 'A', 'C', 'B')
    print("\nPuzzle State per move\n")
    hanoi(disks, A, B, C)
    print("Total moves: ", x)
```

<h4><b>Explanation</b></h4>
<p align="justify">The function tower(n, A, C, B) represents moving n disks from tower A to tower C, using tower B as the helper.
If n is equal to 0, the function stops and returns 0 moves. This is the base case that prevents the program from running forever. When n is greater than 0, the function first moves n-1 disks from tower A to tower B. After that, it prints the move of the biggest disk from tower A to tower C. Finally, it moves the n-1 disks from tower B to tower C.
Each time a disk is moved, the program increases the move counter by 1. The function keeps adding the number of moves from every recursive call and returns the total number of moves at the end. The user is asked to input how many disks they want, and the program checks if the input is valid before running the solution.</p>
