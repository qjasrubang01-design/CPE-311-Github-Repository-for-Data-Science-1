# Question 1

Your goal in this kata is to implement a difference function, which subtracts one list from another and returns the result.

It should remove all values from list a, which are present in list b keeping their order.

    array_diff([1,2],[1]) == [2]
If a value is present in b, all of its occurrences must be removed from the other:

    array_diff([1,2,2,2,3],[2]) == [1,3]

```python
def array_diff(a, b):
    set_b = set(b)
    result = [item for item in a if item not in b]
    return result
```

This function removes all values present in list b from list a while keeping the original order of the remaining items. I solved it by checking if the value from a exists in b then removing that b value if it does.

# Question 2

Your task is to make a function that can take any non-negative integer as an argument and return it with its digits in descending order. Essentially, rearrange the digits to create the highest possible number.

<strong>Examples:</strong>
Input: 42145 Output: 54421

Input: 145263 Output: 654321

Input: 123456789 Output: 987654321

```python
def ips_between(num):
    numbers = list(int (d) for d in str(num))
    numbers.sort( reverse = True)
    result = int("".join(str (num) for num in numbers))
    print(result)
    return result
```

```python
# @title Test code should run and not return error

def test_1():
    assert(ips_between(0)== 0)
def test_2():
    assert(ips_between(15)== 51)
def test_3():
    assert(ips_between(123456789)== 987654321)
test_1()
test_2()
test_3()
```

First, I break the number down into a list of digits and sort them from highest to lowest so that the largest values occupy the highest place positions. Finally, I join these sorted digits back together and convert them into a single integer to produce the maximized result.

For example, for [1, 2, 2] it should return 9 because 1<sup>2</sup>+2<sup>2</sup>+2<sup>2</sup>=9.


```python
def square_sum(numbers):
    #your code here
    squaredSum = 0
    for number in numbers:
        squaredSum+=number**2
    print(squaredSum)
    return squaredSum
```

# Question 4

Write a function that takes in a string of one or more words, and returns the same string, but with all words that have five or more letters reversed (Just like the name of this Kata). Strings passed in will consist of only letters and spaces. Spaces will be included only when more than one word is present.

Examples:

    "Hey fellow warriors"  --> "Hey wollef sroirraw"
    "This is a test        --> "This is a test"
    "This is another test" --> "This is rehtona test"

```python
def spin_words(sentence):
    whole = sentence.split()
    x=[]
    for item in whole:
        words = list(str (d) for d in str(item))
        if len(words) >= 5:
            words.reverse()
            result = "".join(str (letters) for letters in words)
            x.append(result)
        else:
            result = "".join(str (letters) for letters in words)
            x.append(result)
    result = " ".join(str (letters) for letters in x)
    print (result)
    return result
```

```python
# @title Test code should run and not return error

def test_1():
    sample = "Welcome"
    assert(spin_words(sample)) == "emocleW"
def test_2():
    sample = "to"
    assert(spin_words(sample)) == "to"
def test_3():
    sample = "CodeWars"
    assert(spin_words(sample)) == "sraWedoC"
def test_4():
    sample = "Hey fellow warriors"
    assert(spin_words(sample)) == "Hey wollef sroirraw"
def test_5():
    sample = "This sentence is a sentence"
    assert(spin_words(sample)) == "This ecnetnes is a ecnetnes"

test_1()
test_2()
test_3()
test_4()
test_5()
```

I created this method to process a sentence by identifying and reversing any words that contain five or more characters. I start by splitting the string into individual words and then iterate through them, checking the length of each to determine if it needs to be flipped. Once the specific words are reversed, I append everything into a new list and join them back together with spaces to reconstruct the modified sentence.
